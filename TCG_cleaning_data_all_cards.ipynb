```python
import pandas as pd
```

```python
cards = pd.read_csv('karty.csv', sep=',', encoding='latin-1')
```

```python
cards['generation'] = cards['generation'].str.strip()
```

```python
#sprawdzamy, co się dzieje w danych, szukamy zależności, ciekawych rzeczy i czy są gdzieś błędy
all_columns= ['id', 'set', 'series', 'publisher', 'generation', 'release_date',
       'artist', 'name', 'set_num', 'types', 'supertype', 'subtypes', 'level',
       'hp', 'evolvesFrom', 'evolvesTo', 'abilities', 'attacks', 'weaknesses',
       'retreatCost', 'convertedRetreatCost', 'rarity', 'flavorText',
       'nationalPokedexNumbers', 'legalities', 'resistances', 'rules',
       'regulationMark', 'ancientTrait']
print("Ile jest Na w danej kolumnie?")
for column in all_columns:
    print(column," : ",cards[column].isna().sum())
print("\nIle jest wartości danego typu w danej kolumnie?")
for column in all_columns:
    print(column," : ",cards[column].value_counts(),"\n")
```

```python
#usuwam niepotrzebne mi kolumny
cards = cards.drop(columns=[ 'publisher', 'release_date','set','series',
       'artist', 'types', 'subtypes', 'level',
       'hp', 'evolvesFrom', 'evolvesTo', 'abilities', 'attacks', 'weaknesses',
       'retreatCost', 'convertedRetreatCost', 'rarity', 'flavorText',
       'nationalPokedexNumbers', 'resistances', 'rules',
       'regulationMark', 'ancientTrait'])
cards['supertype'] = cards['supertype'].str.replace('PokÃ©mon', 'Pokémon')  #mimo dobrania encoding nadal mam problem z tym słowem więc podmieniam je
```
